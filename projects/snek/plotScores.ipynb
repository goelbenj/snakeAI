```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

scores = pd.read_csv('hamiltonianV2.txt', delimiter='\n')
# scores2 = pd.read_csv('hamiltonianV1.txt', delimiter='\n')
# scores['Scores']
rows = scores.shape[0]
average = scores["Scores"].mean()
print(average)
average_plot = np.ones(rows)*average
# print(average_plot)
# print(rows)
trials = np.linspace(1, rows, rows)
# print(trials)
plt.plot(trials, scores['Scores'], 'r-', markersize=2, label = 'Hamiltonian Path w/ Skips - w/ TIMEOUT')
plt.plot(trials, average_plot, 'b-', markersize=2, label= 'Average')
plt.title('Algorithm Performances vs. Trial #')
plt.xlabel('Trial #')
plt.ylabel('Score')
plt.legend(loc='upper right')
plt.show()
```
